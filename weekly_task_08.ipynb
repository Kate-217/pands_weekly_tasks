Write a program called plottask.py that displays:

* A histogram of a normal distribution of a 1000 values with a mean of 5 and standard deviation of 2
* A plot of the function  h(x)=x^3 in the range 0 to 10, on the one set of axes.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generating random 1000 numbers, which are normally distributed and give a mean of 5 and a standard deviation of 2
number_of_values = 1000 
mean = 5
sd = 2
samples_list =[]
np.random.seed(4)

for i in range(0,10000): # the range wasn't specified, so I put the range from 0 to 10000
    sample = np.random.normal(mean,sd) # the function generates a number and I add it to the list "samples_list"
    samples_list.append(sample)


# Creating a plot of the function  f(x)=x^3
x_values = np.array(range(0,11))
y_values = x_values ** 3


fig, ax = plt.subplots()

# creating a histogram of normal distributions
ax.hist(samples_list,bins = 100,color = 'g',alpha =0.7)
ax.set_title('Normal distribution of a 1000 values with a mean of 5 and standard deviation of 2')
ax.set_xlabel('Values')
ax.set_ylabel('Frequency')

# creating a plot of the function  f(x)=x^3 
ax.plot(x_values, y_values, color = 'r', label= 'f(x)=x^3')
ax.legend()
ax.set_xlim(0, 10)
plt.show()



```
